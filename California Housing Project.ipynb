10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
# Create an empty list to store the categories
total_bedroom_size = []

# Iterate through each value in 'total_bedrooms' and assign the category
for value in Housing['total_bedrooms']:
    if value <= 10:
        total_bedroom_size.append('Small')
    elif 11 <= value < 1000:
        total_bedroom_size.append('Medium')
    else:
        total_bedroom_size.append('Large')

# Assign the list as a new column 'total_bedroom_size'
Housing['total_bedroom_size'] = total_bedroom_size
Housing['total_bedroom_size']
```

Explanation: In this solution, we use a for loop to iterate through each value in the 'total_bedrooms' column. We use conditionals to determine the category for each value and append it to the 'total_bedroom_size' list. Finally, we assign the list as a new column 'total_bedroom_size' in the DataFrame.
